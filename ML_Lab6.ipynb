```python
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import pandas as pd

# Load your data into a DataFrame
data = pd.read_csv('l.csv', encoding='ISO-8859-1')
encoded_data = data.copy()

court_name_encoder = LabelEncoder()
domain_encoder = LabelEncoder()

encoded_data['Court name'] = court_name_encoder.fit_transform(data['Court name'])
encoded_data['Domain'] = domain_encoder.fit_transform(data['Domain'])

# Select the two encoded numerical features
feature1 = encoded_data['Court name']  # Replace with the actual column name
feature2 = encoded_data['Domain']  # Replace with the actual column name

# Create a scatter plot
plt.scatter(feature1, feature2, alpha=0.5)  # alpha controls point transparency

# Set axis labels and title
plt.xlabel('Encoded Court name')
plt.ylabel('Encoded Domain')
plt.title('Scatter Plot of Encoded Court_name vs. Encoded domain')

# Display the plot
plt.show()
```
